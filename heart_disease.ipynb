```python
from sklearn.model_selection import train_test_split

x=dataset.drop('num',axis=1)
y=dataset['num']

X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.20)
```

```python
#logistic regression
```

```python
dataset
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
lr_predict = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(lr_predict,Y_test)
```

```python
#naive Bayes
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,Y_train)
nb_predict = nb.predict(X_test)
```

```python
accuracy_score(nb_predict,Y_test)
```

```python
#SVM
```

```python
from sklearn import svm
sv = svm.SVC(kernel='linear')
sv.fit(X_train,Y_train)
sv_predict = sv.predict(X_test)
```

```python
accuracy_score(sv_predict,Y_test)
```

```python
#KNN
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=30)
knn.fit(X_train,Y_train)
knn_predict=knn.predict(X_test)
```

```python
accuracy_score(knn_predict,Y_test)
```

```python
#decision tree
```

```python
from sklearn.tree import DecisionTreeClassifier

max_accuracy = 0


for x in range(200):
    dt = DecisionTreeClassifier(random_state=x)
    dt.fit(X_train,Y_train)
    Y_pred_dt = dt.predict(X_test)
    current_accuracy = round(accuracy_score(Y_pred_dt,Y_test)*100,2)
    if(current_accuracy>max_accuracy):
        max_accuracy = current_accuracy
        best_x = x
        
#print(max_accuracy)
#print(best_x)


dt = DecisionTreeClassifier(random_state=best_x)
dt.fit(X_train,Y_train)
dt_predict = dt.predict(X_test)
```

```python
accuracy_score(dt_predict,Y_test)
```

```python
#random forest
```

```python
from sklearn.ensemble import RandomForestClassifier

max_accuracy = 0


for x in range(2000):
    rf = RandomForestClassifier(random_state=x)
    rf.fit(X_train,Y_train)
    Y_pred_rf = rf.predict(X_test)
    current_accuracy = round(accuracy_score(Y_pred_rf,Y_test)*100,2)
    if(current_accuracy>max_accuracy):
        max_accuracy = current_accuracy
        best_x = x
        
#print(max_accuracy)
#print(best_x)

rf = RandomForestClassifier(random_state=best_x)
rf.fit(X_train,Y_train)
rf_predict = rf.predict(X_test)
```

```python
print(max_accuracy)
print(best_x)
```

```python
accuracy_score(rf_predict,Y_test)
```

```python
#from all methods Random Forrest gives best result at random_state = 691
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(Y_test,rf_predict))
print("\n")
print(classification_report(Y_test,rf_predict))
```
